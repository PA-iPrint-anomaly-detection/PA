```python
# ==========================================================
# SINGLE MODEL (eventual failure) TRAINED ON MULTI-K (K=1..10)
# + EVALUATION :
#   - For each K: print confusion matrix + Acc/Prec/Rec/F1
#   - For each K: show confusion matrices for multiple test sets together (2x2 grid, paged)
# ==========================================================

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import RobustScaler
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score


# ----------------------------
# CONFIG (paths you provided)
# ----------------------------
FIT_DIR = "/content/drive/MyDrive/data/fit"

# Train on ONE dataset (rep_variant)
TRAIN_DS = "1_lowering"

# Evaluate on multiple other ones (edit as needed)

EVAL_DS = [
    "2_lowering",
    "1_increasing",
]

K_LIST = list(range(1, 11))  # K=1..10
RANDOM_STATE = 42
TEST_SIZE = 0.20

# ----------------------------
# Helpers: paths / sweep order
# ----------------------------
def dataset_key_to_fp(dataset_key: str) -> str:
    rep = int(dataset_key.split("_")[0])
    var = dataset_key.split("_", 1)[1]
    return os.path.join(
        FIT_DIR,
        f"merged_rep_{rep}_{var}_bankcorrected_params6_Q_rmse_nrmse_labels.npz"
    )

def sweep_index(dataset_key: str, pressures: np.ndarray) -> np.ndarray:
    # lowering: sweep high -> low
    # raising/increasing: sweep low -> high
    dk = dataset_key.lower()
    if ("raising" in dk) or ("increasing" in dk):
        return np.argsort(pressures)        # low -> high
    return np.argsort(pressures)[::-1]      # high -> low


# ----------------------------
# Labels: strict jetting/nonjetting only + trim to labeled pressures
# ----------------------------
def strict_labels_jet_nonjet_only(labels_raw: np.ndarray) -> np.ndarray:
    """
    Keep ONLY: 'jetting', 'nonjetting', ''.
    Everything else -> ''.
    """
    s = labels_raw.astype(str)
    s = np.char.strip(s)
    low = np.char.lower(s)

    # normalize blanks
    low[np.isin(low, ["", "nan", "none", "null"])] = ""

    out = np.empty_like(low)
    out[:] = ""

    out[low == "jetting"] = "jetting"

    nonjet_variants = [
        "nonjetting", "non-jetting", "non jetting", "non_jetting",
        "nonjet", "nojet", "no-jetting"
    ]
    out[np.isin(low, nonjet_variants)] = "nonjetting"

    return out


def monotone_filter(labels_slim: np.ndarray, idx_sweep: np.ndarray, base_valid_nozzles: np.ndarray):
    """
    Exclude nozzles that revert NonJet -> Jet AFTER first NonJet (in sweep order).
    Only checks base_valid_nozzles (from file valid_nozzles if present).
    """
    N = labels_slim.shape[0]
    has_revert = np.zeros(N, dtype=bool)

    for i in base_valid_nozzles:
        seq = labels_slim[int(i), idx_sweep]
        is_jet = (seq == "jetting")
        is_non = (seq == "nonjetting")
        if not np.any(is_non):
            continue
        first_non = np.where(is_non)[0][0]
        if np.any(is_jet[first_non + 1:]):
            has_revert[int(i)] = True

    valid_mask = np.zeros(N, dtype=bool)
    valid_mask[base_valid_nozzles] = True
    valid_mask = valid_mask & (~has_revert)

    valid_nozzles = np.where(valid_mask)[0]
    return valid_mask, valid_nozzles, has_revert


def compute_event_pos(labels_slim: np.ndarray, idx_sweep: np.ndarray, valid_nozzles: np.ndarray) -> np.ndarray:
    """
    event_pos[i] = first sweep index where label == nonjetting; -1 if never flips.
    """
    N = labels_slim.shape[0]
    event_pos = np.full(N, -1, dtype=int)

    for i in valid_nozzles:
        seq = labels_slim[int(i), idx_sweep]
        hits = np.where(seq == "nonjetting")[0]
        if hits.size > 0:
            event_pos[int(i)] = int(hits[0])

    return event_pos


def load_dataset_package(dataset_key: str, verbose: bool = True) -> dict:
    fp = dataset_key_to_fp(dataset_key)
    if not os.path.exists(fp):
        raise FileNotFoundError(f"Missing merged NPZ: {fp}")

    m = np.load(fp, allow_pickle=True)

    pressures   = m["pressures"].astype(float)
    labels_raw  = m["labels"]
    params6_all = m["params6_all"] if "params6_all" in m else m["params6"]

    N = params6_all.shape[0]

    # base valid nozzles from file if available
    if "valid_nozzles" in m:
        base_valid_nozzles = np.array(m["valid_nozzles"], dtype=int)
    else:
        base_valid_nozzles = np.arange(N, dtype=int)

    # strict label filtering
    labels_slim = strict_labels_jet_nonjet_only(labels_raw)

    # trim to labeled pressures only (any jet/nonjet anywhere)
    labeled_cols = np.where(np.any(labels_slim != "", axis=0))[0]
    if labeled_cols.size == 0:
        raise ValueError(f"[{dataset_key}] No labeled pressure columns after strict filtering.")

    pressures   = pressures[labeled_cols]
    params6_all = params6_all[:, labeled_cols, :]
    labels_slim = labels_slim[:, labeled_cols]

    # sweep order on trimmed pressures
    idx_sweep = sweep_index(dataset_key, pressures)

    # monotone filter
    valid_mask, valid_nozzles, has_revert = monotone_filter(labels_slim, idx_sweep, base_valid_nozzles)

    # event positions
    event_pos = compute_event_pos(labels_slim, idx_sweep, valid_nozzles)

    if verbose:
        n_valid = int(valid_mask.sum())
        n_onset = int(np.sum((event_pos >= 0) & valid_mask))
        n_revert = int(np.sum(has_revert))
        p0 = float(pressures[idx_sweep[0]])
        p1 = float(pressures[idx_sweep[-1]])
        print(f"[{dataset_key}] P(labeled)={len(pressures)} | sweep {p0:.1f}->{p1:.1f} mBar | "
              f"valid={n_valid} onset={n_onset} revert_excl={n_revert}")

    return dict(
        dataset_key=dataset_key,
        fp=fp,
        pressures=pressures,
        params6_all=params6_all,
        labels_slim=labels_slim,
        idx_sweep=idx_sweep,
        valid_mask=valid_mask,
        valid_nozzles=valid_nozzles,
        event_pos=event_pos,
    )


# ----------------------------
# Feature row (delta) + builders (same logic as your single-K builders)
# ----------------------------
def _row_singleK_delta(nozzle_id: int, y: int, step_pos: int, seq_sweep: np.ndarray,
                      pressures: np.ndarray, idx_sweep: np.ndarray, K: int) -> dict:
    """
    seq_sweep is already in sweep order (P,6).
    Features: d0..d5 + has_prev. (No K/step_pos in features)
    """
    step_pos = int(step_pos)
    x = seq_sweep[step_pos].astype(float)

    has_prev = 1 if step_pos - 1 >= 0 else 0
    if has_prev:
        x_prev = seq_sweep[step_pos - 1].astype(float)
        d = x - x_prev
    else:
        d = np.zeros_like(x)

    p = float(pressures[int(idx_sweep[step_pos])])

    row = {
        "nozzle": int(nozzle_id),
        "y": int(y),
        "K": int(K),                  # metadata only
        "step_pos": int(step_pos),    # metadata only
        "pressure": p,                # metadata only
        "has_prev": int(has_prev),
    }
    for j in range(d.shape[0]):
        row[f"d{j}"] = float(d[j]) if np.isfinite(d[j]) else np.nan
    return row


def build_balanced_matched_singleK_delta(pkg: dict, K: int, rng: np.random.RandomState, subset_nozzles=None):
    params6_all = pkg["params6_all"]
    pressures   = pkg["pressures"]
    idx_sweep   = pkg["idx_sweep"]
    valid_nozz  = pkg["valid_nozzles"] if subset_nozzles is None else np.array(subset_nozzles, dtype=int)
    event_pos   = pkg["event_pos"]

    N, P, _ = params6_all.shape
    if K > P:
        return None, {"status": "K_gt_P"}

    pos_rows = []
    healthy  = []
    skipped_early = 0

    for i in valid_nozz:
        i = int(i)
        T = int(event_pos[i])
        seq = params6_all[i, idx_sweep, :]  # (P,6) in sweep order

        if T >= 0:
            step_pos = T - K
            if step_pos < 0:
                skipped_early += 1
                continue
            pos_rows.append(_row_singleK_delta(i, 1, step_pos, seq, pressures, idx_sweep, K))
        else:
            healthy.append(i)

    df_pos = pd.DataFrame(pos_rows)
    if df_pos.empty:
        return None, {"status": "no_positives", "skipped_early": int(skipped_early)}

    steps = df_pos["step_pos"].to_numpy(int)
    uniq_s, cnt = np.unique(steps, return_counts=True)
    need = dict(zip(uniq_s, cnt))

    healthy = np.array(healthy, dtype=int)
    rng.shuffle(healthy)
    hset = set(map(int, healthy.tolist()))

    pos_used_parts = []
    neg_rows = []

    for s in sorted(need.keys()):
        need_s = int(need[s])
        if not hset:
            break

        available = np.array(sorted(hset), dtype=int)
        take = min(need_s, len(available))
        chosen = rng.choice(available, size=take, replace=False)

        for noz in chosen:
            seq = params6_all[int(noz), idx_sweep, :]
            neg_rows.append(_row_singleK_delta(int(noz), 0, int(s), seq, pressures, idx_sweep, K))

        hset.difference_update(set(map(int, chosen)))
        pos_used_parts.append(df_pos[df_pos["step_pos"] == s].sample(n=take, random_state=RANDOM_STATE))

    if (len(pos_used_parts) == 0) or (len(neg_rows) == 0):
        return None, {"status": "matching_failed", "skipped_early": int(skipped_early)}

    df = pd.concat([pd.concat(pos_used_parts, axis=0), pd.DataFrame(neg_rows)], axis=0)
    df = df.sample(frac=1.0, random_state=RANDOM_STATE).reset_index(drop=True)

    feat_cols = [c for c in df.columns if c.startswith("d")] + ["has_prev"]
    df[feat_cols] = df[feat_cols].replace([np.inf, -np.inf], np.nan)

    meta = dict(
        status="ok",
        n_pos=int((df.y == 1).sum()),
        n_neg=int((df.y == 0).sum()),
        unique_steps=int(df["step_pos"].nunique()),
        skipped_early=int(skipped_early),
    )
    return (df, feat_cols), meta


def build_full_unbalanced_singleK_delta(pkg: dict, K: int, rng: np.random.RandomState,
                                       exclude_nozzles=None, subset_nozzles=None):
    """
    Full eval set (like your notebook):
      - positives: all failing nozzles at t*-K (if valid)
      - negatives: all healthy nozzles, sampled at a step drawn from pos step distribution
    """
    params6_all = pkg["params6_all"]
    pressures   = pkg["pressures"]
    idx_sweep   = pkg["idx_sweep"]
    valid_nozz  = pkg["valid_nozzles"] if subset_nozzles is None else np.array(subset_nozzles, dtype=int)
    event_pos   = pkg["event_pos"]

    N, P, _ = params6_all.shape
    if K > P:
        return None, {"status": "K_gt_P"}

    exclude_nozzles = set() if exclude_nozzles is None else set(map(int, exclude_nozzles))

    pos_rows = []
    pos_steps = []
    healthy = []
    skipped_early = 0

    for i in valid_nozz:
        i = int(i)
        if i in exclude_nozzles:
            continue

        T = int(event_pos[i])
        seq = params6_all[i, idx_sweep, :]

        if T >= 0:
            step_pos = T - K
            if step_pos < 0:
                skipped_early += 1
                continue
            pos_rows.append(_row_singleK_delta(i, 1, step_pos, seq, pressures, idx_sweep, K))
            pos_steps.append(step_pos)
        else:
            healthy.append(i)

    if len(pos_rows) == 0:
        return None, {"status": "no_positives", "skipped_early": int(skipped_early)}

    pos_steps = np.array(pos_steps, dtype=int)

    neg_rows = []
    for i in healthy:
        i = int(i)
        if i in exclude_nozzles:
            continue
        step_pos = int(rng.choice(pos_steps, size=1)[0])
        seq = params6_all[i, idx_sweep, :]
        neg_rows.append(_row_singleK_delta(i, 0, step_pos, seq, pressures, idx_sweep, K))

    df = pd.DataFrame(pos_rows + neg_rows)

    feat_cols = [c for c in df.columns if c.startswith("d")] + ["has_prev"]
    df[feat_cols] = df[feat_cols].replace([np.inf, -np.inf], np.nan)

    meta = dict(
        status="ok",
        n_pos=int((df.y == 1).sum()),
        n_neg=int((df.y == 0).sum()),
        skipped_early=int(skipped_early),
    )
    return (df, feat_cols), meta


# ----------------------------
# Multi-K training set: stack balanced-matched sets across K (ONE model)
# ----------------------------
def build_balanced_matched_multiK_delta(pkg: dict, K_list, train_nozzles: np.ndarray):
    dfs = []
    feat_cols_ref = None
    perK_meta = {}

    for K in K_list:
        rng = np.random.RandomState(RANDOM_STATE + 1000 + K)
        built, meta = build_balanced_matched_singleK_delta(pkg, K, rng, subset_nozzles=train_nozzles)
        perK_meta[int(K)] = meta

        if built is None or meta.get("status") != "ok":
            continue

        dfK, feat_cols = built
        dfK = dfK.copy()
        dfK["K"] = int(K)  # metadata only
        dfs.append(dfK)

        if feat_cols_ref is None:
            feat_cols_ref = feat_cols
        elif feat_cols != feat_cols_ref:
            raise ValueError("Feature columns mismatch across K.")

    if len(dfs) == 0:
        return None, {"status": "no_data_after_all_K", "perK": perK_meta}

    df = pd.concat(dfs, axis=0).sample(frac=1.0, random_state=RANDOM_STATE).reset_index(drop=True)
    return (df, feat_cols_ref), {"status": "ok", "perK": perK_meta, "n_rows": int(len(df))}


# ----------------------------
# Model
# ----------------------------
def make_model():
    return Pipeline(steps=[
        ("impute", SimpleImputer(strategy="median")),
        ("scale", RobustScaler()),
        ("clf", HistGradientBoostingClassifier(
            random_state=RANDOM_STATE,
            learning_rate=0.05,
            max_depth=5,
            max_leaf_nodes=63
        ))
    ])


# ----------------------------
# Evaluation (print + confusion matrices grid per K)
# ----------------------------
def metrics_and_cm(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred, labels=[0, 1])  # [[TN FP],[FN TP]]
    acc  = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, zero_division=0)
    rec  = recall_score(y_true, y_pred, zero_division=0)
    f1   = f1_score(y_true, y_pred, zero_division=0)
    return cm, acc, prec, rec, f1


def print_cm_block(name, split_name, K, cm, acc, prec, rec, f1, n):
    print(f"[{name}] {split_name} | K={K} | n={n} | Acc={acc:.3f} Prec={prec:.3f} Rec={rec:.3f} F1={f1:.3f}")
    print("Confusion matrix [[TN FP],[FN TP]] (rows=true [0,1], cols=pred [0,1]):")
    print(cm)


def plot_confmats_grid(items, suptitle):
    """
    items: list of dict {title, y_true, y_pred}
    show up to 4 in a 2x2 grid, paged outside.
    """
    fig, axes = plt.subplots(2, 2, figsize=(10.6, 7.0))
    axes = axes.ravel()

    for k in range(4):
        ax = axes[k]
        if k >= len(items):
            ax.axis("off")
            continue

        it = items[k]
        cm = confusion_matrix(it["y_true"], it["y_pred"], labels=[0, 1])

        ax.imshow(cm, cmap="Blues")
        ax.set_xticks([0, 1]); ax.set_yticks([0, 1])
        ax.set_xticklabels(["Won't flip", "Will flip"], rotation=20, ha="right")
        ax.set_yticklabels(["Won't flip", "Will flip"])
        ax.set_title(it["title"], fontsize=10)

        for (i, j), v in np.ndenumerate(cm):
            ax.text(j, i, str(v), ha="center", va="center", fontsize=11)

    fig.suptitle(suptitle, fontsize=12)
    fig.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()


# ==========================================================
# RUN
# ==========================================================
# Load packages
pkg_train = load_dataset_package(TRAIN_DS, verbose=True)
pkg_eval = {}
for ds in EVAL_DS:
    try:
        pkg_eval[ds] = load_dataset_package(ds, verbose=True)
    except FileNotFoundError as e:
        print(f"[skip] {ds} missing: {e}")
    except Exception as e:
        print(f"[skip] {ds} load error: {e}")

# Split nozzles on TRAIN dataset (stratified by eventual failure)
valid_nozz = np.array(pkg_train["valid_nozzles"], dtype=int)
y_nozz = (pkg_train["event_pos"][valid_nozz] >= 0).astype(int)

if np.unique(y_nozz).size < 2:
    raise RuntimeError(
        f"[{TRAIN_DS}] Stratified split impossible: only one class present after filters "
        f"(fail_rate={float(np.mean(y_nozz)):.4f}). "
        f"Pick a TRAIN_DS that actually contains failures after strict filtering."
    )

sss = StratifiedShuffleSplit(n_splits=1, test_size=TEST_SIZE, random_state=RANDOM_STATE)
tr_idx, te_idx = next(sss.split(valid_nozz.reshape(-1, 1), y_nozz))
train_nozzles = valid_nozz[tr_idx]
test_nozzles  = valid_nozz[te_idx]

print(f"\n[SPLIT] {TRAIN_DS}: train_nozz={len(train_nozzles)} test_nozz={len(test_nozzles)} | "
      f"train_fail_rate={np.mean(pkg_train['event_pos'][train_nozzles] >= 0):.3f} "
      f"test_fail_rate={np.mean(pkg_train['event_pos'][test_nozzles] >= 0):.3f}")

# Build MULTI-K training set (pooled)
built_multi, meta_multi = build_balanced_matched_multiK_delta(pkg_train, K_LIST, train_nozzles)
if built_multi is None or meta_multi.get("status") != "ok":
    raise RuntimeError(f"Multi-K training build failed: {meta_multi}")

df_train, feat_cols = built_multi
print(f"[TRAIN SET] pooled K={K_LIST[0]}..{K_LIST[-1]} | rows={len(df_train)} | "
      f"pos={(df_train.y==1).sum()} neg={(df_train.y==0).sum()} | features={len(feat_cols)}")

# Train ONE model
model = make_model()
model.fit(df_train[feat_cols], df_train["y"].to_numpy(int))
print("[MODEL] trained once (no K/step_pos used at inference)\n")

# Evaluate per K: confusion matrices for different test sets together
all_rows = []

for K in K_LIST:
    print("\n" + "="*120)
    print(f"[EVAL] K={K} â€” Confusion matrices + metrics across test sets")
    print("="*120)

    items = []

    # 1) TRAIN dataset â€” held-out BALANCED (only held-out nozzles)
    rng0 = np.random.RandomState(RANDOM_STATE + 9000 + K)
    built, meta = build_balanced_matched_singleK_delta(pkg_train, K, rng0, subset_nozzles=test_nozzles)
    if built is None or meta.get("status") != "ok":
        print(f"[{TRAIN_DS}] heldout_balanced | K={K} build failed: {meta.get('status') if meta else 'None'}")
        all_rows.append(dict(Dataset=TRAIN_DS, Split="heldout_balanced", K=int(K), status="build_failed",
                             n=0, acc=np.nan, prec=np.nan, rec=np.nan, f1=np.nan, TN=np.nan, FP=np.nan, FN=np.nan, TP=np.nan))
    else:
        dfK, feat_colsK = built
        y_true = dfK["y"].to_numpy(int)
        y_pred = model.predict(dfK[feat_colsK])
        cm, acc, prec, rec, f1 = metrics_and_cm(y_true, y_pred)
        print_cm_block(TRAIN_DS, "heldout_balanced", K, cm, acc, prec, rec, f1, len(y_true))
        items.append({"title": f"{TRAIN_DS}\nheldout BAL\nAcc={acc:.2f}", "y_true": y_true, "y_pred": y_pred})
        all_rows.append(dict(Dataset=TRAIN_DS, Split="heldout_balanced", K=int(K), status="ok",
                             n=int(len(y_true)), acc=acc, prec=prec, rec=rec, f1=f1,
                             TN=int(cm[0,0]), FP=int(cm[0,1]), FN=int(cm[1,0]), TP=int(cm[1,1])))

    # 2) TRAIN dataset â€” FULL unbalanced excluding training nozzles
    rng1 = np.random.RandomState(RANDOM_STATE + 9100 + K)
    built, meta = build_full_unbalanced_singleK_delta(pkg_train, K, rng1, exclude_nozzles=train_nozzles)
    if built is None or meta.get("status") != "ok":
        print(f"[{TRAIN_DS}] full_excl_train | K={K} build failed: {meta.get('status') if meta else 'None'}")
        all_rows.append(dict(Dataset=TRAIN_DS, Split="full_excl_train", K=int(K), status="build_failed",
                             n=0, acc=np.nan, prec=np.nan, rec=np.nan, f1=np.nan, TN=np.nan, FP=np.nan, FN=np.nan, TP=np.nan))
    else:
        dfK, feat_colsK = built
        y_true = dfK["y"].to_numpy(int)
        y_pred = model.predict(dfK[feat_colsK])
        cm, acc, prec, rec, f1 = metrics_and_cm(y_true, y_pred)
        print_cm_block(TRAIN_DS, "full_excl_train", K, cm, acc, prec, rec, f1, len(y_true))
        items.append({"title": f"{TRAIN_DS}\nFULL excl-train\nAcc={acc:.2f}", "y_true": y_true, "y_pred": y_pred})
        all_rows.append(dict(Dataset=TRAIN_DS, Split="full_excl_train", K=int(K), status="ok",
                             n=int(len(y_true)), acc=acc, prec=prec, rec=rec, f1=f1,
                             TN=int(cm[0,0]), FP=int(cm[0,1]), FN=int(cm[1,0]), TP=int(cm[1,1])))

    # 3) Other datasets â€” FULL unbalanced (all valid nozzles)
    for di, ds in enumerate(EVAL_DS):
        if ds not in pkg_eval:
            continue

        rngE = np.random.RandomState(RANDOM_STATE + 9200 + K + 77*di)
        built, meta = build_full_unbalanced_singleK_delta(pkg_eval[ds], K, rngE, exclude_nozzles=None)
        if built is None or meta.get("status") != "ok":
            print(f"[{ds}] full_unbalanced | K={K} build failed: {meta.get('status') if meta else 'None'}")
            all_rows.append(dict(Dataset=ds, Split="full_unbalanced", K=int(K), status="build_failed",
                                 n=0, acc=np.nan, prec=np.nan, rec=np.nan, f1=np.nan, TN=np.nan, FP=np.nan, FN=np.nan, TP=np.nan))
            continue

        dfK, feat_colsK = built
        y_true = dfK["y"].to_numpy(int)
        y_pred = model.predict(dfK[feat_colsK])
        cm, acc, prec, rec, f1 = metrics_and_cm(y_true, y_pred)
        print_cm_block(ds, "full_unbalanced", K, cm, acc, prec, rec, f1, len(y_true))

        items.append({"title": f"{ds}\nFULL\nAcc={acc:.2f}", "y_true": y_true, "y_pred": y_pred})
        all_rows.append(dict(Dataset=ds, Split="full_unbalanced", K=int(K), status="ok",
                             n=int(len(y_true)), acc=acc, prec=prec, rec=rec, f1=f1,
                             TN=int(cm[0,0]), FP=int(cm[0,1]), FN=int(cm[1,0]), TP=int(cm[1,1])))

    # Plot confusion matrices (2x2), paged if >4
    for page in range(0, len(items), 4):
        chunk = items[page:page+4]
        plot_confmats_grid(
            chunk,
            suptitle=f"K={K} â€” Confusion matrices (0=Won't flip, 1=Will flip)  [page {page//4 + 1}]"
        )

# Summary table
df_summary = pd.DataFrame(all_rows).sort_values(["Dataset", "Split", "K"]).reset_index(drop=True)

print("\n" + "="*120)
print("SUMMARY (all datasets, all K)")
print("="*120)
print(df_summary.to_string(index=False))

out_csv = "single_model_metrics_confmats_perK.csv"
df_summary.to_csv(out_csv, index=False)
print(f"\n[saved] {out_csv}")
```
